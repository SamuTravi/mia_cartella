```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
from cmath import *

def MakeImage(F, n, scale=0.01):
    data = [scale*i for i in range(-n,n)]
    return np.matrix([[F(complex(a, b)) for a in data] for b in data])

def DrawImage(F, n, scale):
    m = MakeImage(F, n, scale)
    plt.figure(figsize=(16,16))
    img = plt.imshow(m, extent=(scale*n, -scale*n, scale*n, -scale*n), cmap='hot')
    plt.colorbar()
    plt.show()

    
def JuliaSetRec(z, c, k, max_k=64):
    while k<max_k:
        if abs(z)>2:
            return k
        z=z**2+c
        k=k+1            
    return 0
        
        

def JuliaSet(z, c=-0.413):
    return JuliaSetRec(0, z, 0)       # ho cambiato qua (0,z,0) invece che (z,c,0)  

DrawImage(JuliaSet, 450, 0.005)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
from cmath import *

def MakeImage(F, n, scale=0.01):
    data = [scale*i for i in range(-2*n,5*n)]
    return np.matrix([[F(complex(a, b)) for a in data] for b in data])

def DrawImage(F, n, scale):
    m = MakeImage(F, n, scale)
    plt.figure(figsize=(16,16))
    img = plt.imshow(m, extent=(scale*n, -scale*n, scale*n, -scale*n), cmap='hot')
    plt.colorbar()
    plt.show()

    
def JuliaSetRec(z, c, k, max_k=64):       # si possono anche aumentare le iterazioni (con poche iterazioni c' contrasto)
    while k<max_k:
        if abs(z)>2:
            return k
        z=z**2+c
        k=k+1            
    return 0
        
        

def JuliaSet(z, c=-0.413):
    return JuliaSetRec(0, z, 0)       # ho cambiato qua (0,z,0) invece che (z,c,0)  

DrawImage(JuliaSet, 100, 0.0007)        # il primo  per la risoluzione, il secondo  per lo zoom
```
